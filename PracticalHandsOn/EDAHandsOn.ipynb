```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('tested.csv')
df.head()
```

# Univariate plots 

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

### Count Plots

```python
df['Pclass'].value_counts().plot(kind='bar')
plt.title('Number of passengers by Pclass')
plt.xlabel('Pclass')
plt.ylabel('Number of passengers')
plt.show()
# Using seaborn
sns.countplot(x='Pclass', data=df)
```

```python
df['Pclass'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Number of passengers by Pclass')
# seaborn does not have a pie chart function
```
